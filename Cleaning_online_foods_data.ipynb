```python
"""Data Cleaning List
Deleting redundant columns,
Renaming the columns,
Dropping duplicates,
Remove the NaN values from the dataset,
Check for some more transformation
"""
```

```python
data.head(1) #See available header
```

```python
data.columns
```

```python
columns_to_keep = ['Age', 'Gender', 'Marital Status', 'Occupation', 'Monthly Income',
       'Educational Qualifications', 'Family size'] #Grouping which column wnat to keep
```

```python
columns_to_drop = ['latitude', 'longitude',
       'Pin code', 'Unnamed: 12'] #Group which column want to drop
```

```python
len(columns_to_keep)
```

```python
len(columns_to_drop)
```

# 1st method - filtering the columns that we wanted

```python
modified_data = data[columns_to_keep]
```

```python
modified_data
```

# 2nd method -  Dropping unnecessary columns 

```python
data.drop(columns = columns_to_drop, inplace=True)
```

```python
data # Here dataset after we drop unwanted column
```

# Rename columns

```python
data.head(1)
```

```python
data.columns
```

```python
data.rename(columns ={'Monthly Income': 'Monthly Salary'}, inplace= True) #Change from 'Montly Income' to 'Monthly Salary'
```

```python
data
```

# Dropping the duplicates

```python
data.duplicated().value_counts() # Here show that there 144 duplicate row available in the dataset
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
# Since from here the sum is 0, means that there no duplicates left.
```

# Remove NaN values from dataset

```python
data.isna().sum()
# Since there no NaN value available , mean it clean from NaN values
```

```python
data
```

# How do we know how many month does it total counted from age column each students?

```python
data['Age (Months)'] = data['Age']* 12
```

```python
data
```

```python
# Change  'Gender' column Male to M and  Female to F
data['Gender'] = data['Gender'].apply(lambda x: "F" if x == "Female" else "M" )
```

```python
data
```

```python
data.head(3)
```

# Grouping family size into 'Large' & 'Small'.

```python
# If family size > 3 then its become Large, else consider as small.

data['Family size'] = data['Family size']. apply(lambda x: 'Large' if x > 3 else 'Small' )
```

```python
data
```

```python
# Change Output from Yes as 1 and No as 0
data['Output'] = data['Output'].apply(lambda x: 1 if x == 'Yes' else '0')
```

```python
# changes data type from float boolean to float
data['Output'] = data['Output'].astype(float)
```

```python
type(data['Output'][1])
```

# Save the file into CSV format.

```python
data.to_csv('Cleaned_onlinefoods_data.csv', index=False)
```
